```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Create and fit the DecisionTreeClassifier
model_all_params = DecisionTreeClassifier().fit(iris.data, iris.target)

# Prepare a plot figure with set size.
plt.figure(figsize=(20, 10))

# Plot the decision tree, showing the decisive values and the improvements in Gini impurity along the way.
plot_tree(model_all_params, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)

# Display the plot
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Create and fit the DecisionTreeClassifier
model_all_params = DecisionTreeClassifier().fit(iris.data, iris.target)

# Prepare a plot figure with set size.
plt.figure(figsize=(20, 10))

# Plot the decision tree, showing the decisive values and the improvements in Gini impurity along the way.
plot_tree(model_all_params, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)

# Display the plot
plt.show()
```

```python
model_all_params_min_imp_dec_001 = DecisionTreeClassifier(min_impurity_decrease = 0.01).fit(iris.data, iris.target)
# Prepare a plot figure with set size.
plt.figure(figsize = (16,10))
# Plot the decision tree.
plot_tree(model_all_params_min_imp_dec_001,
          rounded = True,
          filled = True                )
# Display the tree plot figure.
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Create and fit multiple DecisionTreeClassifier models with different parameters
models = []
for max_depth in range(3, 10):
    model = DecisionTreeClassifier(max_depth=max_depth)
    model.fit(iris.data, iris.target)
    models.append(model)

# Prepare a plot figure with set size.
plt.figure(figsize=(20, 12))

# Plot the decision tree from the model at index 5, showing the decisive values and the improvements in Gini impurity along the way.
plot_tree(models[5], rounded=True, filled=True)

# Display the plot
plt.show()
```
